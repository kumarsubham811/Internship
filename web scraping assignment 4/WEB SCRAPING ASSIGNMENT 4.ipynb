# Q1

```python
#question 1
driver1=webdriver.Chrome()
```

```python
#step 1: get the link
driver1.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#step 2: scraping the data
#rank can't be scraped form the site
rank=[]
rank1=driver1.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr')
for i in rank1:
    rank.append(i.get_attribute[':before'])
rank
```

```python
#so i have create the loop for rank as the scraped data in in the order of rank
rank=[]
x=0
for i in range(30):
    if x<30:
        x=x+1
    else:
        break
    rank.append(x)
rank
```

```python
#scraping name
name=[]
name1=driver1.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in name1:
    name.append(i.text)
name
```

```python
#scraping artist
artist=[]
artist1=driver1.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for i in artist1:
    artist.append(i.text)
artist
```

```python
#scraping views
views=[]
views1=driver1.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for i in views1:
    views.append(i.text)
views
```

```python
#scraping date
date=[]
date1=driver1.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for i in date1:
    date.append(i.text)
date
```

```python
#step 3: insert the scraped data into a dataframe 
import pandas as pd
df1=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'No. of views':views,'Upload date':date})
df1
```

```python
#question 4
driver4=webdriver.Chrome()
```

```python
#step1: get the link
driver4.get('https://github.com/')
```

```python
 driver4.maximize_window()
```

```python
#step 2: click on open source button
open_source=driver4.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
open_source.click()
```

```python
#step 3: click on trending button
trending=driver4.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
trending.click()
```

```python
#step 4: scraping the data
#scraping repository title
title=[]
title1=driver4.find_elements(By.XPATH,'//article[@class="Box-row"]/h2/a')
for i in title1:
    title.append(i.text)
title
```

```python
len(title)
```

```python
#scraping url
url=[]
url1=driver4.find_elements(By.XPATH,'//article[@class="Box-row"]/h2/a')
for i in url1:
    url.append(i.get_attribute('href'))
url
```

```python
#scraping repository description, contributers count , language used
from selenium.common.exceptions import NoSuchElementException
description=[]
count=[]
language=[]
for i in url:
    driver4.get(i)
    time.sleep(5)
    try:
        describe=driver4.find_element(By.XPATH,'//p[@class="f4 my-3"]')
        description.append(describe.text)
    except NoSuchElementException:
        description.append('-')
    
    try:
        count1=driver4.find_element(By.XPATH,'//span[@class="Counter ml-1"]')
        count.append(count1.text)
    except NoSuchElementException:
        count.append('-')
    
    try:
        l1=driver4.find_element(By.XPATH,'//span[@class="color-fg-default text-bold mr-1"]')
        language.append(l1.text)
    except NoSuchElementException:
        language.append('-')
description
```

```python
#contributers count is having a problem as it has differnt xpaths for contributer count and some of them is not interacting properly as well
#so i'm creating the dataframe with the data it's able to scrape
count2=[]
for i in url[0:6]:
    driver4.get(i)
    time.sleep(5)
    try:
        count1=driver4.find_elements(By.XPATH,'//div[@class="Layout-sidebar"]/div/div[4]/div/h2/a/span')
        for p in count1:                     
            count2.append(p.text)
    except NoSuchElementException:
        count1=driver4.find_elements(By.XPATH,'//div[@class="Layout-sidebar"]/div/div[3]/div/h2/a/span')
        for p in count1:                     
            count2.append(p.text)
    except NoSuchElementException:
        count1=driver4.find_elements(By.XPATH,'//div[@class="Layout-sidebar"]/div/div[5]/div/h2/a/span')
        for p in count1:                     
            count2.append(p.text)
    except NoSuchElementException:                     
        count2.append('-')
count2
```

```python
count
```

```python
language
```

```python
#step 5: creating a dataframe and putting scraped data into it
import pandas as pd
df4=pd.DataFrame({'Repository title':title,'Repository description':description,'Contributors count':count,'Language used':language})
df4
```
